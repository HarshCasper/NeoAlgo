# Implementing a simple decision tree classifier in python

```python
#First import the necessary python packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import data using the pandas package
#here we are taking the iris dataset
df = pd.read_csv('../input/iris/Iris.csv') 
```

```python
#To view the data
df.head() 
```

```python
#Extract features and separate the target variable
X=df.drop("Species",axis=1)
y=df["Species"]
```

```python
#printing data type and shape 
print(type(X))
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
 #Splitting the dataset into train and test
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,stratify=y,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#Creating the classifier object 
model_gini = DecisionTreeClassifier(criterion = "gini",random_state = 42,max_depth=3, min_samples_leaf=5)
```

```python
#Training model with gini index
model_gini.fit(X_train,y_train)
```

```python
#Predicting on test set
model_gini.predict(X_test)
```

```python
#Evaluating the model
model_gini.score(X_test,y_test)
```
